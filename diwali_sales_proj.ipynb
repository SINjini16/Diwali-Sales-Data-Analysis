```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis = 1, inplace =True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.describe()
```

```python
df[['Age','Marital_Status', 'Orders']].describe()
```

```python
ax = sns.countplot(x='Gender', data=df)
```

```python
x = sns.countplot(x='Gender',data=df)

for bars in x.containers:
  x.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
sns.set(rc={'figure.figsize':(5,10)})
sales_by_gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.pie(sales_by_gender['Amount'], labels=sales_by_gender['Gender'], autopct='%.2f%%', startangle=90)
plt.title('Sales by Gender')
plt.show()
```

```python
sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
sns.set(rc={'figure.figsize':(10,15)})
sales_by_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.pie(sales_by_age['Amount'], labels=sales_by_age['Age Group'], autopct='%.0f%%', startangle=90)
plt.title('Sales by Age Group')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(5,10)})
sales_by_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
plt.pie(sales_by_age['Amount'], labels=sales_by_age['Age Group'], autopct='%.0f%%', startangle=90)
plt.title('Sales by Age Group')
plt.show()
```
